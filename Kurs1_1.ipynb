```python
n = int(input())
a = [[0] * n for i in range(n)]
i0, ik, j0, jk = 0, len(a), 0, len(a[0])
x = 1
isSecond = False
isDone = False
while 1:
    if isDone:
        break
        
    if isSecond:
        j0 += 1
        
    for j in range(j0, jk):
        a[i0][j] = x
        x += 1
        if x > n * n:
            isDone = True
            break
            
    if isDone:
        break
    
        
    for i in range(i0 + 1, ik):
        a[i][jk - 1] = x
        x += 1
        if x > n * n:
            isDone = True
            break
            
    if isDone:
        break
        
    for j in range(jk - 2, j0 - 1, -1):
        a[ik - 1][j] = x
        x += 1
        if x > n * n:
            isDone = True
            break
            
    i0 += 1
    
    if isDone:
        break
    for i in range(ik - 2, i0 - 1, -1):
        a[i][j0] = x
        x += 1
        if x > n * n:
            isDone = True
            break
            
    isSecond = True
    ik -= 1
    jk -= 1

for i in range(len(a)):
    for j in range(len(a[0])):
        print(a[i][j], end = ' ')
    print()
```

```python
def f(x):
    f = 0
    if x <= -2:
        f = 1 - (x + 2) ** 2
    elif x > -2 and x <= 2:
        f = -x / 2
    elif x > 2:
        f = (x - 2) ** 2 + 1
    return f
```

```python
def modify_list(l):
    helpArr = []
    for x in l:
        if x % 2 == 0:
            helpArr += [x // 2]
    l.clear()
    for x in helpArr:
        l.append(x)
lst = [1, 2, 3, 4, 5, 6]
print(modify_list(lst))  # None
print(lst)               # [1, 2, 3]
modify_list(lst)
print(lst)               # [1]

lst = [10, 5, 8, 3]
modify_list(lst)
print(lst)               # [5, 4]
```

```python
def update_dictionary(d, key, value):
    helpD = d.copy()
    if helpD.get(key) != None:
        helpD[key] += [value]
    elif helpD.get(2 * key) != None:
        helpD[2 * key] += [value]
    else:
        helpD[2 * key] = [value]
    d.update(helpD)
d = {}
print(update_dictionary(d, 1, -1))  # None
print(d)                            # {2: [-1]}
update_dictionary(d, 2, -2)
print(d)                            # {2: [-1, -2]}
update_dictionary(d, 1, -3)
print(d)                            # {2: [-1, -2, -3]}
```
